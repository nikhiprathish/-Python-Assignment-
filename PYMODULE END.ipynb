```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Replace 'your_dataset.csv' with the actual file path
df = pd.read_csv('myexcel.csv')
```

```python
df
```

```python
# Replacing 'Height' values with random numbers between 150 and 180
df['Height'] = np.random.randint(150, 181, size=len(df))
```

```python
df
```

```python
team_distribution = df['Team'].value_counts()
team_percentage = (team_distribution / len(df)) * 100

# Display the results
print("Team Distribution:\n", team_distribution)
print("\nTeam Percentage:\n", team_percentage)
```

```python
position_distribution = df['Position'].value_counts()
print("Position Distribution:\n", position_distribution)
```

```python
# Define age bins
bins = [20, 25, 30, 35, 40, 45, 50]
labels = ['20-25', '25-30', '30-35', '35-40', '40-45', '45-50']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Find predominant age group
age_group_distribution = df['Age_Group'].value_counts()
print("Age Group Distribution:\n", age_group_distribution)
```

```python
# Sum of salary expenditure by team
salary_by_team = df.groupby('Team')['Salary'].sum()

# Sum of salary expenditure by position
salary_by_position = df.groupby('Position')['Salary'].sum()

# Find the maximum for each
highest_salary_team = salary_by_team.idxmax()
highest_salary_position = salary_by_position.idxmax()

print("Team with Highest Salary Expenditure:", highest_salary_team)
print("Position with Highest Salary Expenditure:", highest_salary_position)
```

```python
# Calculate correlation
correlation = df[['Age', 'Salary']].corr().iloc[0, 1]
print(f"Correlation between Age and Salary: {correlation}")

# Plot correlation
sns.scatterplot(data=df, x='Age', y='Salary')
plt.title('Correlation between Age and Salary')
plt.show()
```

```python
team_distribution.plot(kind='bar', color='skyblue')
plt.title('Employee Distribution Across Teams')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show()
```

```python
position_distribution.plot(kind='bar', color='lightgreen')
plt.title('Employee Distribution Across Positions')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.show()
```

```python
age_group_distribution.plot(kind='bar', color='blue')
plt.title('Employee Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
salary_by_team.plot(kind='bar', color='gray')
plt.title('Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.show()
```

```python
salary_by_position.plot(kind='bar', color='pink')
plt.title('Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.show()
```
